```python
import pandas as pd
```

```python
# csv 파일 열기
df = pd.read_csv('score.csv')
df
```

```python
# 1번째 줄을 제외하고 가져옴.
df = pd.read_csv('score.csv',skiprows=1)
df
```

```python
# 1,3,5번째 줄을 제외
df = pd.read_csv('score.csv',skiprows=[1,3,5])
df
```

```python
# 1,2,3번째 줄을 제외 (1번째는 index임)
df = pd.read_csv('score.csv',skiprows=[0,1,2])
df
```

```python
# 두 줄을 제외 (1번째는 index임)
df = pd.read_csv('score.csv',skiprows=2)
df
```

```python
# 지정된 개수 만큼 출력
# index는 존재하고 아래에 있는 데이터 4개줄을 가져옴.
df = pd.read_csv('score.csv',nrows=4)
df
```

```python
# 처음 2줄을 제외, 4개 데이터를 가져오기
df = pd.read_csv('score.csv',skiprows=2,nrows=4)
df
```

```python
# txt파일 열기
df = pd.read_csv('score.txt')
df
```

```python
# \t 구분자를 사용해서 가져오기
df = pd.read_csv('score.txt',sep='\t')
df
```

```python
# index를 지정해서 가져오기
df = pd.read_csv('score.txt',index_col='지원번호')
df
```

```python
# 불러와서 index 지정하는 방법
df = pd.read_csv('score.txt')
df.set_index('지원번호',inplace=True)
df
```

```python
# excel 파일 불러오기
df = pd.read_excel('score.xlsx')
df
```

```python
# index 지정해서 엑셀 파일 불러오기
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
# 엑셀 파일 불러온 다음에 index 지정
df = pd.read_excel('score.xlsx')
df.set_index('지원번호',inplace=True)
df
```
