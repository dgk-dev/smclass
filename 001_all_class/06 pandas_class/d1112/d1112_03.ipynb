```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1,2,3]
y = [2,4,8]
# 선그래프 생성
plt.plot(x,y)
# 상단 타이틀
plt.title("라인그래프")
# 그래프 X축설명, color:글자색변경
plt.xlabel('X축 날짜',color='#a1d3ff',loc='left') # 그래프하단 축설명
plt.ylabel('Y축 인원',loc='top')

# 그래프 눈금 단위 표시
plt.xticks([1,2,3])
plt.yticks([2,4,8])

plt.show()
```
