```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm

# 한글 폰트 설정 (나눔고딕 사용)
font_path = r'C:\Windows\Fonts\malgun.ttf'  # 맑은 고딕 폰트 경로
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# 그래프 스타일 및 크기 설정
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(12, 6))

# 데이터 준비
names = df['이름']
subjects = ['국어', '영어', '수학']
scores = [df['국어'], df['영어'], df['수학']]
colors = ['#FF9999', '#66B2FF', '#99FF99']

# 누적 막대 그래프 생성
bottom = np.zeros_like(df['국어'])
bars = []
for subject, score, color in zip(subjects, scores, colors):
    bar = plt.bar(names, score, bottom=bottom, width=0.6,
                  label=subject, color=color, alpha=0.8)
    bars.append(bar)
    bottom += score

# 그래프 스타일링
plt.title('학생별 과목 성적 분포', pad=20, fontsize=14, fontweight='bold')
plt.xlabel('학생 이름', labelpad=10, fontsize=12)
plt.ylabel('점수', labelpad=10, fontsize=12)

# 격자 추가
plt.grid(axis='y', linestyle='--', alpha=0.3)

# y축 범위 및 눈금 설정
plt.ylim(0, 300)
plt.yticks(np.arange(0, 301, 50))

# 점수 텍스트 표시
for i in range(len(names)):
    # 국어 점수
    plt.text(i, df['국어'][i]/2, f'{df["국어"][i]}',
             ha='center', va='center', color='white', fontweight='bold')
    
    # 영어 점수
    plt.text(i, df['국어'][i] + df['영어'][i]/2, f'{df["영어"][i]}',
             ha='center', va='center', color='white', fontweight='bold')
    
    # 수학 점수
    plt.text(i, df['국어'][i] + df['영어'][i] + df['수학'][i]/2, f'{df["수학"][i]}',
             ha='center', va='center', color='white', fontweight='bold')
    
    # 총점 표시
    total = df['국어'][i] + df['영어'][i] + df['수학'][i]
    plt.text(i, total + 5, f'총점: {total}',
             ha='center', va='bottom', color='darkred', fontweight='bold')

# 범례 설정
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')

# 여백 조정
plt.tight_layout()

plt.show()
```
