```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Win용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
## 원그래프
values = [30,25,20,13,10,2]
explode = [0.3,0.1,0,0,0,0]
```

```python
labels = ['python','java','javascript','oracle','html','etc']
plt.title('프로그램 사용률')
## pie함수(데이터값), labels: 글자출력
## explode : 원그래프 분리, countclock : 시계반대방향, False : 시계방향

plt.pie(values,labels=labels,explode=explode,autopct='%.1f%%',counterclock=False,startangle=90)
plt.legend(loc=(1.2,0.3))
plt.show()
```
