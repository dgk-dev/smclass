```python
import pandas as pd
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
```

```python
### 영화 관객수를 막대그래프로 출력하시오.
### 평균 막대그래프를 추가하고, 평균 선을 그어서... 표시하시오.

df.info()
```

```python
num_avg = df['관객 수'].mean()
rate_avg = df['평점'].mean()
df.loc[8] = ['평균',2024,num_avg, rate_avg]

# 그래프 크기 설정
plt.figure(figsize=(15,7))

# 막대 그래프 생성
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc', '#99ccff']
bars = plt.bar(df['영화'], df['관객 수'], color=colors, width=0.5)

plt.axhline(y=num_avg, color='red', linestyle='--', alpha=0.7, label=f'평균: {num_avg:.0f}')

plt.title('영화 관객수')
plt.xlabel('영화 이름')
plt.ylabel('관객수(단위: 만)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.0f}',
             ha='center', va='bottom')


plt.tight_layout()
```
