```python
###
# 1. 요일 컬럼 추가
# 2. 0-6까지의 숫자를 랜덤으로 입력
# 3. 함수를 사용해서 0-일요일, 1-월요일, 2-화요일..... 6-토요일로 변경하시오.

# 1. 요일 컬럼 추가하고 랜덤 숫자 생성
import numpy as np
df['요일'] = np.random.randint(0, 7, size=len(df))

# 2. 숫자를 요일로 변환하는 함수 정의
def number_to_weekday(x):
    weekdays = {
        0: '일요일',
        1: '월요일',
        2: '화요일',
        3: '수요일',
        4: '목요일',
        5: '금요일',
        6: '토요일'
    }
    return weekdays[x]

# 3. 함수를 적용하여 숫자를 요일로 변환
df['요일'] = df['요일'].apply(number_to_weekday)
```
