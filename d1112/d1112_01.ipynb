```python
### 모든 술의 총량을 출력하시오.
### total_servings / 전체 알코올 도수를 확인
# 전체 알코올 도수 확인



### 
```

```python
# 데이터 불러오기
df = pd.read_csv('drinks.csv')

# 1. 각 술의 총량 계산
total_beer = df['beer_servings'].sum()
total_spirit = df['spirit_servings'].sum()
total_wine = df['wine_servings'].sum()

# 2. 전체 알코올 도수 합계
total_alcohol = df['total_litres_of_pure_alcohol'].sum()

print("술 종류별 총 소비량:")
print(f"맥주 총량: {total_beer:,} 잔")
print(f"증류주 총량: {total_spirit:,} 잔")
print(f"와인 총량: {total_wine:,} 잔")
print(f"\n순수 알코올 총량: {total_alcohol:.2f} 리터")

# 3. 전체 주류 소비량
total_servings = total_beer + total_spirit + total_wine
print(f"\n전체 주류 소비량: {total_servings:,} 잔")
```

```python
### 대륙별 total_servings을 구하시오.

# 1. 먼저 각 술의 소비량을 합쳐서 total_servings 컬럼 생성
df['total_servings'] = df['beer_servings'] + df['spirit_servings'] + df['wine_servings']

# 2. 대륙별로 그룹화하여 합계 계산
continent_total = df.groupby('continent')['total_servings'].sum()

# 3. 결과 출력 (내림차순 정렬)
continent_total = continent_total.sort_values(ascending=False)
print(continent_total)
```

```python
### 술 소비량 대비 알콜 비율이 어느 정도 되는지 출력하시오.
### 알코올 비율을 계산해서 alcohol_rate 컬럼을 추가
### total_litres_of_pure_alcohol 알코올 비율을 total_servings 으로 나눈 것
### 술소비량 대비 알콜 비율이 높은 나라별 20개를 출력하시오.
### 대륙별 알콜비율도 계산해서 출력하시오.


# 1. total_servings 컬럼 생성 (전체 술 소비량)
df['total_servings'] = df['beer_servings'] + df['spirit_servings'] + df['wine_servings']

# 2. alcohol_rate 컬럼 생성 (알코올 비율)
df['alcohol_rate'] = df['total_litres_of_pure_alcohol'] / df['total_servings']
# 0으로 나누는 것을 방지
df['alcohol_rate'] = df['alcohol_rate'].replace([np.inf, -np.inf], np.nan)

# 3. 알코올 비율이 높은 상위 20개 국가
top_20_countries = df.sort_values('alcohol_rate', ascending=False).head(20)
print("알코올 비율이 높은 상위 20개 국가:")
print(top_20_countries[['country', 'alcohol_rate']])

# 4. 대륙별 알코올 비율 계산
continent_alcohol = df.groupby('continent').agg({
    'total_servings': 'sum',
    'total_litres_of_pure_alcohol': 'sum'
}).reset_index()

continent_alcohol['alcohol_rate'] = (continent_alcohol['total_litres_of_pure_alcohol'] / 
                                   continent_alcohol['total_servings'])

print("\n대륙별 알코올 비율:")
print(continent_alcohol.sort_values('alcohol_rate', ascending=False))


```
