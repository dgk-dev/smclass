```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False

import pandas as pd
import random
```

```python
x = [1,2,3]  # 리스트, 튜플, DataFrame 데이터 - 리스트형태, Series
y = [2,4,8]  # SM 매출액
y2 = [5,1,3] # LG 매출액
y3 = [1,2,5] # SK 매출액
plt.figure(figsize=(10,5))
plt.plot(x,y,label='SM매출액',marker='o')
plt.plot(x,y2,label='LG매출액',marker='x')
plt.plot(x,y3,label='SK매출액',marker='>')

plt.title("기업별 매출액")
plt.legend(ncol=2)   # 범례출력 기본: 세로출력, n
plt.show()
```
