```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager, rc

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # 윈도우의 기본 한글 폰트
plt.rcParams['axes.unicode_minus'] = False     # 마이너스 기호 깨짐 방지

# 그래프 및 폰트 크기 설정
plt.figure(figsize=(10, 6))
plt.rcParams['font.size'] = 12

# 데이터 준비
names = df['이름']
subjects = ['국어', '영어', '수학']
scores = [df['국어'], df['영어'], df['수학']]

# 누적 값 계산
bottoms = np.zeros_like(df['국어'])
colors = ['#3498db', '#e67e22', '#2ecc71']

# 바 차트 생성
for subject, score, color in zip(subjects, scores, colors):
    plt.bar(names, score, bottom=bottoms, label=subject, 
            color=color, alpha=0.8)
    bottoms += score

# 그래프 꾸미기
plt.title('학생별 과목 성적', pad=20, fontsize=14)
plt.xlabel('학생 이름', labelpad=10)
plt.ylabel('점수', labelpad=10)

# x축 레이블 설정
plt.xticks(rotation=45, ha='right')

# 범례 설정
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# 여백 조정
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 그래프 크기 설정
plt.figure(figsize=(12, 6))

# 데이터 준비
names = df['이름']
subjects = ['국어', '영어', '수학']
scores = [df['국어'], df['영어'], df['수학']]

# 막대 차트 설정
bar_width = 0.2  # 더 얇은 막대 너비
x = np.arange(len(names))
colors = ['#3498db', '#e67e22', '#2ecc71']

# 각 과목별 막대 그래프 생성
for i, (subject, score, color) in enumerate(zip(subjects, scores, colors)):
    plt.bar(x + i * bar_width, 
            score, 
            width=bar_width, 
            label=subject, 
            color=color,
            alpha=0.8)

# 그래프 스타일링
plt.title('학생별 과목 성적', pad=20, fontsize=14)
plt.xlabel('학생 이름', labelpad=10)
plt.ylabel('점수', labelpad=10)

# x축 레이블 위치 조정
plt.xticks(x + bar_width, names)

# 격자 추가
plt.grid(axis='y', linestyle='--', alpha=0.3)

# y축 범위 설정
plt.ylim(0, 100)

# 범례 위치 조정
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')

# 여백 자동 조정
plt.tight_layout()

plt.show()
```

```python
# drink.csv 파일로
# beer_servings,spirit_servings,wine_servings 합계
# total_servings 컬럼추가
# country별 total_servings 막대그래프 생성

import pandas as pd
import matplotlib.pyplot as plt

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 데이터 읽기
df = pd.read_csv('drinks.csv')

# total_servings 컬럼 추가 
df['total_servings'] = df['beer_servings'] + df['spirit_servings'] + df['wine_servings']

# total_servings 기준으로 상위 50개국 정렬
top_50 = df.nlargest(50, 'total_servings')

# 그래프 크기 설정
plt.figure(figsize=(15, 10))

# 막대 그래프 생성
bars = plt.bar(top_50['country'], top_50['total_servings'])

# 그래프 스타일링
plt.title('국가별 주류 소비량 Top 50', pad=20, fontsize=10)
plt.xlabel('국가', labelpad=10)
plt.ylabel('총 소비량(Servings)', labelpad=10)

# x축 레이블 회전
plt.xticks(rotation=45, ha='right')

# 격자 추가
plt.grid(axis='y', linestyle='-', alpha=1)

# 각 막대 위에 값 표시
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height):,}',
             ha='center', va='bottom')

# 여백 조정
plt.tight_layout()

plt.show()
```
