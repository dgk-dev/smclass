```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Win용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 메뉴별 주문개수
# 순서 상관없이, head(10)로 10개 막대그래프를 출력하시오.)
# 평균치도 계산해서 표현하고 선을 만들어서 표현하세요., 치폴레에 대해서 사람들이 더 자세히 알 수 있는 데이터를 재량껏 추가해보세요.
```

```python
# 데이터 읽기
df = pd.read_csv('chipotle.tsv', sep='\t')

# 메뉴별 주문 개수 계산
item_counts = df['item_name'].value_counts().head(5)

# 그래프 크기 설정
plt.figure(figsize=(10, 6))

# 막대 그래프 생성
bars = plt.bar(item_counts.index, item_counts.values, 
               color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc'])

# 그래프 스타일링
plt.title('상위 5개 메뉴 주문 횟수', pad=20, fontsize=14)
plt.xlabel('메뉴명', labelpad=10)
plt.ylabel('주문 횟수', labelpad=10)

# x축 레이블 회전
plt.xticks(rotation=45, ha='right')

# 격자 추가
plt.grid(axis='y', linestyle='--', alpha=0.3)

# 각 막대 위에 값 표시
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height):,}',
             ha='center', va='bottom')

# 여백 조정
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 데이터 읽기
df = pd.read_csv('chipotle.tsv', sep='\t')

# 메뉴별 주문 개수 계산
menu_count = df['item_name'].value_counts().head(10)
menu_avg = df['item_name'].value_counts().mean()

# 그래프 크기 설정
plt.figure(figsize=(15, 8))

# 막대 그래프 생성
bars = plt.bar(range(len(menu_count)), menu_count.values, 
               color=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FF99CC',
                     '#99CCFF', '#FFB366', '#FF99FF', '#99FFCC', '#FFB3B3'])

# 평균선 추가
plt.axhline(y=menu_avg, color='red', linestyle='--', alpha=0.7,
            label=f'평균 주문량: {menu_avg:.1f}회')

# 그래프 스타일링
plt.title('Chipotle 메뉴별 주문 횟수 TOP 10', pad=20, fontsize=15)
plt.xlabel('메뉴명', labelpad=10, fontsize=12)
plt.ylabel('주문 횟수', labelpad=10, fontsize=12)

# x축 레이블 설정
plt.xticks(range(len(menu_count)), menu_count.index, rotation=45, ha='right')

# 격자 추가
plt.grid(axis='y', linestyle='--', alpha=0.3)

# 각 막대 위에 값 표시
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height):,}회',
             ha='center', va='bottom')

# 추가 정보 표시
info_text = (
    '주요 통계:\n'
    f'• 총 주문 건수: {len(df["order_id"].unique()):,}건\n'
    f'• 평균 주문 금액: ${df["item_price"].str.replace("$","").astype(float).mean():.2f}\n'
    f'• 가장 인기있는 메뉴: {menu_count.index[0]}\n'
    f'• 전체 메뉴 종류: {len(df["item_name"].unique())}개'
)

plt.text(1.15, 0.95, info_text,
         transform=plt.gca().transAxes,
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'),
         fontsize=10,
         verticalalignment='top')

# 범례 표시
plt.legend()

# 여백 조정
plt.tight_layout()

plt.show()
```
